````python
import numpy as np
import matplotlib.pyplot as plt

# Define the 2D array
array_2d = [
    [13, 13, 12, 12, 12, 12, 11, 11],
    [13, 12, 12, 12, 11, 11, 11, 10],
    [12, 12, 8, 7, 6, 5, 10, 10],
    [12, 12, 7, 6, 5, 4, 10, 10],
    [12, 11, 6, 5, 4, 3, 10, 9],
    [11, 11, 5, 4, 3, 2, 9, 9],
    [11, 11, 10, 10, 10, 9, 9, 9],
    [11, 10, 10, 10, 9, 9, 9, 8]
]

# Determine the maximum width of the rows
max_width = max(len(row) for row in array_2d)

# Normalize the array to have the same number of columns in each row by padding with 0
normalized_array = [row + [0] * (max_width - len(row)) for row in array_2d]

# Convert to numpy array for visualization
array_np = np.array(normalized_array)

# Display the array as an image
plt.figure(figsize=(10, 8))
plt.imshow(array_np, cmap='viridis', interpolation='nearest')

# Add text annotations
for i in range(len(array_2d)):
    for j in range(len(array_2d[i])):
        plt.text(j, i, str(array_2d[i][j]), ha='center', va='center', color='white')

plt.colorbar()
plt.title("2D Array Visualization")
plt.axis('off')
plt.show()
# ``` &#8203;:citation[oaicite:0]{index=0}&#8203;
````

````python
import numpy as np

# Function to dynamically rescale grayscale values
def dynamic_rescale(gray_value, min_gray, max_gray):
    """
    Rescales the grayscale value to the full 0-255 range.
    
    Parameters:
    gray_value (int): The grayscale value to rescale.
    min_gray (int): The minimum grayscale value in the original range.
    max_gray (int): The maximum grayscale value in the original range.
    
    Returns:
    int: The rescaled grayscale value.
    """
    return int((gray_value - min_gray) * 255 / (max_gray - min_gray))

# Example usage:
min_gray = 0    # Minimum grayscale value in the original range
max_gray = 255  # Maximum grayscale value in the original range

# Generate the LUT (Look Up Table)
LUT = np.array([dynamic_rescale(i, min_gray, max_gray) for i in range(256)])

# Print the LUT
print(LUT)
````
